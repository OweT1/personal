# Q2
Generate a new column "SMA_5", which calculates the 5-trading-day Simple Moving Average of each companies' stock (for a stock ticker, at the end of each trading day, take the average of the past 4 trading days' "PRICECLOSE" along with that day's "PRICECLOSE"). Note that it is expected that the first 4 trading days of each ticker to not have a defined "SMA_5". (12.5 points)

## Approach:
Use the `.rolling()` function to calculate the 5-trading-day moving average. We note that since this rolling average is for each stock ticker, we will need to "partition" by the `TICKER` column.

```python
df["SMA_5"] = df.groupby("TICKER")["PRICECLOSE"].transform(lambda x: x.rolling(5).mean())

print(df.head())

# Checking to see that SMA_5 are NaN for the first 4 dates of each stock
print(df[df["SMA_5"].isna()].head(10))
```

# Q3
Generate the "DAILY_RETURN", which calculates the 1-day return for each stock (in percentage) at each trading day. The 1-day return of today is calculated based on the "PRICECLOSE" of the most recent trading day against the "PRICECLOSE" of its previous trading day. Express values as a percentage, e.g. a 21.0% return should have the value 21.0 (12.5 points)

## Approach:
We can simply "lag" the `PRICECLOSE` column by using the `.shift()` function. We still need to take note of the partitioning between the various stocks.

```python
df["PREV_CLOSE"] = df.groupby("TICKER")["PRICECLOSE"].transform(lambda x: x.shift(1))
df["DAILY_RETURN"] = round((df["PRICECLOSE"] - df["PREV_CLOSE"])/df["PREV_CLOSE"] * 100, 1)
df = df.drop(["PREV_CLOSE"], axis = 1)

print(df.head())

# Checking that all the DAILY_RETURN are NaN for all the first dates of each stock
print(df[df["DAILY_RETURN"].isna()].head())
```

# Q4
Generate the "MONTHLY_RETURN", which calculates the 30-day return for each stock (in percentage) at each trading day. Note that this is 30 calendar days, where non-trading days take on the values of their last trading day. Similar to Q3, reference prices from the "PRICECLOSE" column. Express values as a percentage, e.g. a 21.0% return should have the value 21.0 (12.5 points)


## Approach:
We will first find the date that is 30 days before the date in each row. We will then check to see if that date is in the data (>= 4/1/21) and is a trading day. With this date, we can then find the index of which this date is represented in the data for each stock, and thereafter extract the `PRICECLOSE` for the date.

```python
df["NEW_DATE"] = pd.to_datetime(df["DATE"], format = "%d/%m/%y") # convert to datetime format
#df["NEW_DATE"] = [dt.datetime(year = int("20" + date.split("/")[2]), month = int(date.split("/")[1]), day = int(date.split("/")[0])) for date in df["DATE"]]
df["DATE_30DAYBEFORE"] = [date + relativedelta(days = -30) for date in df["NEW_DATE"]] # get the "30th day before" for the date in each row

print(df.head())
```

```python
print(df["DATE_30DAYBEFORE"])
```

```python
start_date = min(df["NEW_DATE"]) # 2021-01-04
end_date = max(df["NEW_DATE"]) # 2021-12-31

# nyse = mcal.get_calendar('NYSE')
# trading_days = nyse.valid_days(start_date=start_date, end_date=end_date)

lost_dates = sorted(list(filter(lambda x: x >= df["NEW_DATE"][0], set(df["DATE_30DAYBEFORE"]).difference(set(df["NEW_DATE"]))))) # dates that are not in the date range (ie earlier than 4/1/21 or not a trading day)

# We will create a mapping to map the "30th day before" to the latest trading day
lost_dates_mapping = {}

for lost_date in lost_dates:
  diff = -1
  while lost_date not in lost_dates_mapping:
    new_date = lost_date + relativedelta(days = diff)
    if new_date in list(df["NEW_DATE"]):
      lost_dates_mapping[lost_date] = new_date
    else:
      diff -= 1

print(lost_dates_mapping)
```

```python
# We will map the new dates to a new column
df["DATE_30DAYBEFORE_NEW"] = pd.Series(map(lambda x: lost_dates_mapping[x] if x in lost_dates_mapping else x, df["DATE_30DAYBEFORE"]))

df[df["DATE"] == "23/2/21"].head() # 2021-01-24 is a Sunday, the new date, 2021-01-22, is a Friday
```

```python
#df["DATE_30DAYBEFORE_NEW_IND"] = df.groupby("TICKER")["DATE_30DAYBEFORE_NEW"].transform(lambda x: df[df["NEW_DATE"] == x].index.tolist())
# df_copy = df
# df["DATE_30DAYBEFORE_NEW_IND"] = [df[(row["TICKER"] == new_df["TICKER"] & row["DATE_30DAYBEFORE_NEW"] == new_df["NEW_DATE"]) for row in df.rows].index]
```

```python
# We will get the index of which the 30th day before for each stock exists in the data. If it does not exist, we will replace it with NaN
def get_index(date, ticker):
    return df[(df['NEW_DATE'] == date) & (df['TICKER'] == ticker)].index

df["DATE_30DAYBEFORE_NEW_IND"] = df.apply(lambda row: get_index(row['DATE_30DAYBEFORE_NEW'], row['TICKER']).to_list(), axis=1)
df["DATE_30DAYBEFORE_NEW_IND"] = [int(ind[0]) if ind != [] else np.nan for ind in df["DATE_30DAYBEFORE_NEW_IND"]]
df[df["DATE"] == "23/2/21"].head()
```

```python
# We will use the index to extract the `PRICECLOSE` for the respective dates
df["PRICE_30DAYBEFORE"] = [df["PRICECLOSE"][ind] if not np.isnan(ind) else np.nan for ind in df["DATE_30DAYBEFORE_NEW_IND"]]

# Checking of `PRICECLOSE`
print(df[df["DATE"] == "23/2/21"].head())
print(df[df["DATE"] == "22/1/21"].head())
```

```python
df["MONTHLY_RETURN"] = round((df["PRICECLOSE"] - df["PRICE_30DAYBEFORE"])/df["PRICE_30DAYBEFORE"] * 100, 1)

print(df[df["DATE"] == "23/2/21"].head())
```

```python
# Dropping of columns introduced earlier
new_columns = ["NEW_DATE", "DATE_30DAYBEFORE", "DATE_30DAYBEFORE_NEW", "DATE_30DAYBEFORE_NEW_IND", "PRICE_30DAYBEFORE"]
df = df.drop(new_columns, axis = 1)

print(df.head())
print(df.shape)
```
